8. Create a function that takes in a `string` and removes all non-alphanumeric characters.

```python
def remove_non_alphanum(string):
    for char in string:
        if not char.isalnum():
            string = string.replace(char,"")
    return(string)

remove_non_alphanum("helo*&wor2l")
            
```

10. Create a function that prints the fibonacci sequence from 0 to 25. 

```python
def fibonacci(cap):
    sequence=[]
    #print(type(sequence))
    sequence.append(0)
    sequence.append(1)
    x=1
    while x < cap-1:
        x=x+1
        sequence.append(sequence[x-1]+sequence[x-2])

    return(sequence)

fibonacci(9)
```

What is a static method anyway? 


```python
class actress():
    def __init__(self, name = 'Meryl Streep'):
        self.name = name
        
    def act(self, line):
        print(f'{self.name}: {line}')
        
    def finale():
        return 'The end'
        
meryl = actress()
meryl.act('This is my line...')
actress.finale()
```

```python
#A static method does not take in a self or class parameter, so it cannot modify the class or object state. 
```
